```python
locations = pd.read_csv('elden-ring-data/locations.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

region_counts = locations['region'].value_counts().reset_index()
region_counts.columns = ['region', 'count']

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 7))
ax = sns.barplot(
    data=region_counts,
    x='region',
    y='count',
    palette='pastel',
    edgecolor='black'
)

for i, row in region_counts.iterrows():
    ax.text(i, row['count'] + 0.5, str(row['count']), ha='center', va='bottom', fontsize=12, weight='semibold')

plt.title('Number of Locations per Region', fontsize=16, weight='bold')
plt.xlabel('')
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=30, fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```
